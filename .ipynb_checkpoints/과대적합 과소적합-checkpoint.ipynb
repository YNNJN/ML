```python
import numpy as np
import pandas as pd
class SingleLayer:

  def __init__(self, learning_rate=0.1):
      self.w = None
      self.b = None
      self.losses = []
      self.w_history = []
      self.lr = learning_rate

  def forpass(self, x):
    z = np.sum(x*self.w) + self.b # 가중치와 입력 특성의 곱을 모두 더하기 위해 np.sum()함수 사용
    return z

  def backprop(self, x, err):
    w_grad = x*err #가중치에 대한 그레디언트를 계산함
    b_grad = 1*err #절편에 대한 그레디언트를 계산함
    return w_grad, b_grad

  def fit(self, x, y, epochs=100):
    self.w = np.ones(x.shape[1])                          #가중치 초기화
    self.b = 0                                            #절편 초기화
    self.w_history.append(self.w.copy( ))                 #가중치 기록
    np.random.seed(42)                                    #무작위 시드 지정
    for i in range(epochs):                               #epochs 만큼 반복
      loss = 0
      indexes = np.random.permutation(np.arange(len(x)))  #index random
      for i in indexes:                                   #모든 샘플에 대해 반복
        z = self.forpass(x[i])                            #정방향 계산
        a = self.activation(z)                            #활성화 함수 적용
        err = -(y[i] - a)                                 #오차 계산
        w_grad, b_grad = self.backprop(x[i], err)         #절편 업데이트
        self.w -= self.lr * w_grad                        #가중치 업데이트(학습률 적용)
        self.b -= b_grad                                  #절편 업데이트
        self.w_history.append(self.w.copy( ))             #가중치 초기화
        a = np.clip(a, 1e-10, 1-1e-10)                    #클리핑 후 손실 누적
        loss += -(y[i]*np.log(a)+(1-y[i]*np.log(1-a)))  
      self.losses.append(loss/len(y))                     #에포크 마다 평균손실

  def activation(self,z):
    a = 1 / (1 + np.exp(-z)) #시그모이드 계산
    return a

  def predict(self, x):
    z = [self.forpass(x_i) for x_i in x]  #선형함수 적용
    return np.array(z) > 0
  
  def score(self, x, y):
    return np.mean(self.predict(x) == y)
```

# 적절한 편향-분산 트레이드 오프
편향 줄이면(훈련세트 성능 높임) 분산 커짐(검증세트와 성능 차이 커짐)<br>
분산 줄이면 편향 커짐<br>
적절한 정도를 찾아야함..

```python
class SingleLayer:
  def __init__(self, learning_rate=0.1):
      self.w = None
      self.b = None
      self.losses = []
      self.val_losses = []
      self.w_history = []
      self.lr = learning_rate  

  def activation(self,z):
    a = 1 / (1 + np.exp(-z)) #시그모이드 계산
    return a

  def predict(self, x):
    z = [self.forpass(x_i) for x_i in x]  #선형함수 적용
    return np.array(z) > 0
  
  def score(self, x, y):
    return np.mean(self.predict(x) == y)

  def forpass(self, x):
    z = np.sum(x*self.w) + self.b # 가중치와 입력 특성의 곱을 모두 더하기 위해 np.sum()함수 사용
    return z

  def backprop(self, x, err):
    w_grad = x*err #가중치에 대한 그레디언트를 계산함
    b_grad = 1*err #절편에 대한 그레디언트를 계산함
    return w_grad, b_grad

  def fit(self, x, y, epochs=100, x_val=None, y_val=None):
    self.w = np.ones(x.shape[1])                          #가중치 초기화
    self.b = 0                                            #절편 초기화
    self.w_history.append(self.w.copy( ))                 #가중치 기록
    np.random.seed(42)                                    #무작위 시드 지정
    for i in range(epochs):
      loss = 0
      indexes = np.random.permutation(np.arange(len(x)))  #index random
      for i in indexes:                                   #모든 샘플에 대해 반복
        z = self.forpass(x[i])                            #정방향 계산
        a = self.activation(z)                            #활성화 함수 적용
        err = -(y[i] - a)                                 #오차 계산
        w_grad, b_grad = self.backprop(x[i], err)         #절편 업데이트
        self.w -= self.lr * w_grad                        #가중치 업데이트(학습률 적용)
        self.b -= b_grad                                  #절편 업데이트
        self.w_history.append(self.w.copy( ))             #가중치 초기화
        a = np.clip(a, 1e-10, 1-1e-10)                    #클리핑 후 손실 누적
        loss += -(y[i]*np.log(a)+(1-y[i]*np.log(1-a)))  
      self.losses.append(loss/len(y))                     #에포크 마다 평균손실
      self.update_val_loss(x_val, y_val)                  #검증세트 전달받을 매개변수 추가

  #훈련 세트 손실 계산하는 방식과 동일
  def update_val_loss(self, x_val, y_val):
    if x_val is None:
      return
    val_loss = 0
    for i in range(len(x_val)):                                     #정방향 계산
      z = self.forpass(x_val[i])                                    #활성화 함수 적용
      a = self.activation(z)
      a = np.clip(a, 1e-10, 1-1e-10)
      val_loss += -(y_val[i]*np.log(a)+(1-y_val[i]*np.log(1-a)))
    self.val_losses.append(val_loss/len(y_val))                     #1에포크마다 1번씩 호출


```

# 과대적합 규제

규제 방법
1. 가중치 규제(L1,L2 규제)
    - 가중치의 값 커지지 않도록 제한
2. Early Stopping
  - 검증집합 오류가 최저인 지점에서 멈춤.
  - ex) p = 10 => 10번이상 오류가 검증오류가   - 예전보다 크면 조기멈춤 
3. Data Augmentation
  - 데이터 확대
  - 충분이 큰 training set로 학습
    - 선형변환
    - 비선형 변환
4. dropout 드롭아웃
    - 뉴런의 연결을 임의로 삭제하는 것이다.
5. 배치 정규화
  - 표준편차를 적절히 설정하지 못하면 활성화값이 분포가 잘 되지 않고 치우칠 수 있음.
  - Batch Normalization은 여기서 활성화값이 적당히 퍼지도록 강제하는 방법이다.

# 1. 가중치 규제


1.L1 규제(가중치의 절댓값인 L1추가)

손실함수 + L1 = L1규제
image.png

α = 하이퍼 파라미터
α크면 = 전체 손실함수의 값이 커지지않도록 w(벡터) 값의 합 작아져야함 => 규제가 강해졌다고 한다(가중치가 작아졌으므로)

α작아지면 = 전체 손실함수의 값이 큰 폭으로 커지지 않음. => 규제가 약해졌다고 한다(가중치가 작아졌으므로)

L1 규제 미분

image.png

w_grad += alpha * np.sign(w)

alpha = 규제 하이퍼파라미터

라쏘모델 =  회귀모델에 L1규제 추가한 것 라쏘모델이라 함.

(사이킷런에서 라쏘모델 제공중)


단점 : 하이퍼파라미터에 많은 의존 => 가중치의 크기에 따라 규제의 양이 변하지 않으므로 규제 효과 좋지않음

2. L2규제


image.png

w_grad += alpha * w

회귀모델 + L2규제 => 릿지모델



image.png

```python
#로지스틱 회귀 규제 적용

  def __init__(self, learning_rate=0.1 l1=0, l2=0): #default = 0
      self.w = None
      self.b = None
      self.losses = []
      self.val_losses = []
      self.w_history = []
      self.lr = learning_rate  

      self.l1 = l1                                   #매개변수 추가
      self.l2 = l2

  def fit(self, x, y, epochs=100, x_val=None, y_val=None):
    self.w = np.ones(x.shape[1])                          
    self.b = 0                                            
    self.w_history.append(self.w.copy( ))                 
    np.random.seed(42)                                    
    for i in range(epochs):
      loss = 0
      indexes = np.random.permutation(np.arange(len(x)))  
      for i in indexes:                                   
        z = self.forpass(x[i])                            
        a = self.activation(z)                            
        err = -(y[i] - a)                                 
        w_grad, b_grad = self.backprop(x[i], err)         

        #그레이디언트에서 패널티 항의 미분값 더함
        w_grad += self.l1 * np.sign(self.w) + self.l2 * self.w

        self.w -= self.lr * w_grad                        
        self.b -= b_grad                                  
        self.w_history.append(self.w.copy( ))             
        a = np.clip(a, 1e-10, 1-1e-10)                    
        loss += -(y[i]*np.log(a)+(1-y[i]*np.log(1-a)))  
      self.losses.append(loss/len(y))                     
      self.update_val_loss(x_val, y_val)                  

#update_val_loss에서 호출
  def reg_loss(self):
    return self.l1 * np.sum(np.abs(self.w)) + self.l2 / 2 * np.sum(self.w**2)
```
