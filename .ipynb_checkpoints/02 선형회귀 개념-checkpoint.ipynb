- 아래 그래프에서 주황색 선이나 초록색 선에 있는 값들이 y^이라 불리는 예측값

## 예측값으로 올바른 모델 찾기

1. 무작위로 w와 b를 정합니다.(무작위로 모델 만들기)
2. x에서 샘플 하나를 선택하여 y^을 계산합니다.(무작위로 모델 예측하기)
3. y^와 선택한 샘플의 진짜 y를 비교합니다. (예측한 값과 정답 비교하기, 틀릴 확률 99%)
4. y^이 y와 더 가까워지도록 w,b를 조정합니다.(모델 조정하기)
5. 모든 샘플을 처리할 때까지 다시 2~4를 반복

## 1. 단순히 따라해보기

1. 무작위로 w와 b 초기화

```python
w=1.0
b=1.0
```

2. 샘플 하나(첫번째 샘플 x[0])를 선택하여 y^ 계산

```python
y_hat=x[0]*w+b
print(y_hat)
```

3. 진짜 y와 예측값 y_hat 비교

```python
print(y[0])
```

4. w값 조정(w를 0.1만큼 증가)

```python
w_inc=w+0.1
y_hat_inc=x[0]*w_inc+b
print(y_hat_inc)
```

y_hat보다 조금 증가함

5. w 값 조정한 후 예측값 증가 정도 확인
- y_hat이 증가한 양을 w가 증가한 양으로 나누면 됨
- 첫번째 훈련데이터 x[0]에 대한 w의 변화율

```python
w_rate=(y_hat_inc-y_hat)/(w_inc-w)
print(w_rate)
```

신기하게도 w_rate에 대한 코드를 수식으로 적어 정리하면 변화율은 결국 훈련데이터의 첫번째 샘플인 x[0]이라는 것을 알 수 있음

image.png

- 위의 경우 y_hat의 값은 y보다 작으므로 y_hat의 값을 증가시켜야 함.
- 이때 변화율은 양수이므로 w값을 증가시키면 y_hat의 값을 증가시킬 수 있다.
- 만약 변화율이 음수이면 w값을 감소시키면 됨

### 하지만 이 방법은 변화율이 음수일 때와 양수일 때를 구분해야 해서 번거로움-> 효율적 해결 방법은?

## 변화율로 가중치 업데이트 하기

### 1. 변화율이 양수일 때 가중치를 업데이트 하는 방법
- 아래 그래프는 변화율이 0보다 클 때를 가정한 그래프
- 목표는 y_hat 값을 증가시키는 것
- 변화율이 양수인 점을 이용하여 변화율을 w에 더하는 방법으로 w 증가시킴

image.png

### 2. 변화율이 음수일 때 가중치를 업데이트 하는 방법
- 이 상황에서는 w가 감소해야 y_hat이 증가
- 변화율이 음수이므로 w에다가 더해주면 y_hat을 증가시킬 수 있음


image.png

# **위의 두 경우 모두 w에 변화율을 더하면 된다는 것을 알았음(w+w_rate)**


```python
w_new=w+w_rate
print(w_new)
```

## 변화율로 절편 업데이트 하기

절편 b에 대한 변화율을 구한 다음 변화율로 b를 업데이트 하기(b를 0.1만큼 증가)

```python
b_inc=b+0.1
y_hat_inc=x[0]*w+b_inc
print(y_hat_inc)
```

```python
b_rate=(y_hat_inc-y_hat)/(b_inc-b)
print(b_rate)
```

- 변화율은 1
- 즉, b가 1만큼 증가하면 y_hat도 1만큼 증가

b를 업데이트 하기 위해서는 변화율이 1이므로 단순히 1을 더하면 됨

```python
b_new=b+1
print(b_new)
```

위 방법으로는 아래 상황들에 적절하게 대처하기 어려움
- y_hat이 y에 한참 미치지 못하는 값인 경우, w와 b를 더 큰 폭으로 수정할 수 없음
- y_hat이 y보다 커지면 y_hat을 감소시키지 못함

더 능동적인 업데이트 방법인 오차역전파

## 오차가 클수록 더 많이 변화시켜야 함 -> 그걸 반영하려면 위처럼 변화율을 더할 때 (오차*변화율)을 더해주면 됨

위의 상황들에 대처 가능
- 필요한만큼 w와 b를 더 크게 수정할 수 있음
- y^이 y를 지나치면 방향을 바꿔줌

1. 오차와 변화율을 곱하여 가중치 업데이트 하기

```python
err=y[0]-y_hat
w_new=w+w_rate*err
b_new=b+1*err
print(w_new,b_new)
```

2. 두번째 샘플 x[1]을 사용하여 오차를 구하고 새로운 w와 b를 구함
- 앞에서 w_rate 식을 정리했을 때, 샘플값과 같아진다는 것을 알았으므로 w_rate을 별도로 계산하지 않고 샘플값을 그대로 사용
- 즉 여기서는 x[1]이 w_rate가 됨

```python
y_hat=x[1]*w_new+b_new
err=y[1]-y_hat
w_rate=x[1]
w_new=w+w_rate*err
b_new=b+1*err
print(w_new,b_new)
```

3. 전체 샘플을 반복하기

```python
for x_i,y_i in zip(x,y):
  y_hat=x_i*w+b
  err=y_i-y_hat
  w_rate=x_i
  w=w+w_rate*err
  b=b+1*err
print(w,b)  
```

* zip함수는 여러개의 배열에서 동시에 요소를 하나씩 꺼내줌

4. 과정 3을 통해 얻어낸 모델이 데이터 세트를 잘 표현하는지 그래프를 그려 알아봄
- 직선 그래프를 그리려면 시작점과 종료점의 x좌표와 y좌표를 plot()함수에 전달하면 됨

```python
plt.scatter(x,y)
pt1=(-0.1,-0.1*w+b)
pt2=(0.15,0.15*w+b)
plt.plot([pt1[0],pt2[0]],[pt1[1],pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

5. 여러 에포크를 반복하기
- 보통 경사하강법에서는 주어진 훈련데이터로 학습을 여러번 반복
- 전체 훈련데이터를 모두 이용하여 한단위의 작업을 진행하는 것을 에포크라고 부름
- 100번 반복하여 직선이 어떻게 이동하는지 확인

```python
for i in range(1,100):
  for x_i,y_i in zip(x,y):
    y_hat=x_i*w+b
    err=y_i-y_hat
    w_rate=x_i
    w=w+w_rate*err
    b=b+1*err
print(w,b)
```

```python
plt.scatter(x,y)
pt1=(-0.1,-0.1*w+b)
pt2=(0.15,0.15*w+b)
plt.plot([pt1[0],pt2[0]],[pt1[1],pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

x가 0.18일 때 y^의 값 예측해보기
