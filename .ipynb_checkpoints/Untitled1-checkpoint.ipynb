# 분류 문제

## 이진 분류

- 임의의 샘플 데이터를 True / False로 구분함

## 퍼셉트론 (Perceptron)

- 이진 분류 문제에서 최적의 가중치를 학습함
- 계단함수까지 통과한 값을 다시 가중치와 절편을 구하기 위한 학습에 사용

#### 뉴런과 비교할 때 퍼셉트론의 인자 수가 더 많음
- 뉴런은 노드(w,x,b)가 하나
- 퍼셉트론은 인풋(w1x1, w2x2, ... , b)이 여러개 -> sigma로 표현 가능
    - z = b + sigma(w_i,x_i)
    - 위와 같이 여러 개의 특성을 식으로 표기할 수 있음
    - w와 x의 곱으로 표현한 이유는, 곱으로 표현해야 선형 표현이 가능하기 때문


## 아달린 (Adaline)

- 선형함수의 결과를 가중치와 절편을 구하는 학습에 사용하고 계단 함수의 결과는 예측에만 활용하는 적응형 선형 뉴런임
- 퍼셉트론은 계단함수의 결과를 학습에 사용하므로, 아달린과 구분됨

## 로지스틱 회귀 (Logistic regression)

- 과정이 다음과 같음
    - z가 선형함수를 통과함
    - z를 **활성화함수**를 통과시켜 a로 만듦
    - a의 데이터를 학습에 사용하고
    - 임계 함수를 통해 이진 분류 함

#### 활성화함수 내용 추가

- 활성화함수는 주로 비선형 함수를 사용함
- 선형함수의 활성화함수를 사용할 경우, 결국 다시 하나의 큰 선형함수가 되어버리기 때문
- 망을 깊게 만들고, 데이터 은닉화를 이루기 위해 비선형 함수를 사용함
    - 은닉층은 신경망의 여러 층을 의미
    - 은닉층에서 활성화함수를 태울 수도 있고, 은닉층을 벗어나서 출력층 바로 앞에서 사용할 수도 있음
- 이를 통해 레이어를 깊게 쌓아도 결과값을 우리가 원하는 꼴로 맞춰줄 수 있음



## 시그모이드 함수 (Sigmoid function)

- 로지스틱 회귀의 활성화함수
- 오즈 비 > 로짓함수 > 시그모이드 함수의 과정으로 만들어짐

1. 오즈 비(Odds ratio): 성공확률과 실패확률의 비율

    - OR(odds ratio) = p/(1-p) (p = 성공확률)
    - 즉 성공확률 / 실패확률

2. 로짓함수(Logit function): 오즈비에 로그를 취한 함수

    - logit(p) = log_e(p/(1-p))
    - p가 0.5일 때 0, prk 0,1일 때 각각 무한대로 음수와 양수가 됨



3. 로지스틱 함수(시그모이드 함수): 로짓함수를 z에 대해 정리한 것 (z를 가로축에 쓰기 위해)

    - p = 1/(1+e^-z) (최종)
    - 유도 과정은 아래와 같음
    - log_e(p/(1-p)) = z
    - p/(1-p) = e^z
    - p(1 + e^z) = e^z
    - p = e^z / (1 + e^z) = 1 / (1+e^-z)

**로지스틱 회귀는 결국 0과 1로 이진분류를 하고자 함**

- 손실함수의 목표는 정답의 예상값의 오차제곱이 최소가 되는 가중치와 절편을 찾는 것이고
- 분류의 목표는 올바르게 분류된 샘플 데이터의 비율 자체를 높이는 것(정확도 up)이니
- 로지스틱 손실함수를 경사하강법에 적용하자



- 분류된 샘플은 미분 가능한 함수가 아니기 때문에 경사하강법을 위한 손실함수로 사용이 불가함
- 손실함수로 사용하기 위해 로지스틱 손실 함수를 설정해야
    - L = -(ylog(a) + (1-y)log(1-a))

- 체인룰에 따라 1) a에 대해 편미분 2) z에 대해 편미분 3) w에 대해 편미분하는 과정을 거침
- 가중치 미분, 절편 모두 동일한 과정을 거침
