# 경사 하강법 실습

1. w와 b 초기화하기<br>
w와 b를 무작위로 초기화한다.

```python
w = 3.1
b = 5.4
```

2. 훈련 데이터의 첫 번째 샘플 데이터로 ŷ 얻기


```python
y_hat = x[0] * w + b
print(y_hat)
```

3. 타깃과 예측 데이터 비교하기

```python
print(y[0])
```

4. w 값 조절해 예측값 바꾸기<br>
w와 b를 조금씩 변경해서 y_hat이 증가하는지 또는 감소하는지 살펴본다.

```python
w_inc = w + 0.1
y_hat_inc = x[0] * w_inc + b
print(y_hat_inc)
```

5. w 값 조정한 후 예측값 증가 정도 확인하기<br>
y_hat이 증가한 양을 w가 증가한 양으로 나누면 된다.<br>
w_rate에 대한 코드를 수식으로 적어 정리하면 변화율은 결국 x[0]이다.

```python
w_rate = (y_hat_inc - y_hat) / (w_inc - w)
print(w_rate)
print(x[0])
```

&nbsp;

#### 변화율에 따른 가중치 업데이트 방법
* 변화율이 양수일 때 y_hat을 증가시키려면 w를 증가시켜야한다.
* 변화율이 음수일 때 y_hat을 증가시키려면 w를 감소시켜야한다.

```python
w_new = w + w_rate
print(w_new)
```

&nbsp;

#### 변화율로 절편 업데이트하기
절편 b에 대한 변화율을 구한 다음 변화율 b를 업데이트
b를 업데이트하기 위해서는 변화율이 1이므로 단순히 1을 더하면 된다.

```python
b_inc = b + 0.1
y_hat_inc = x[0] * w + b_inc
print(y_hat_inc)
```

```python
b_rate = (y_hat_inc - y_hat) / (b_inc - b)
print(b_rate)
```

* 문제점
  * y_hat이 y에 한참 미치지 못하는 값인 경우 w와 b를 더 큰 폭으로 수정할 수 없다.
  * y_hat이 y보다 커지면 y_hat을 감소시키지 못한다.

&nbsp;

#### 오차 역전파로 가중치와 절편 업데이트하기

y에서 ŷ를 뺀 오차의 양을 변화율에 곱하는 방법으로 w를 업데이트<br>
ŷ이 y보다 많이 작은 경우 w와 b를 많이 바꿀 수 있다.<br>
ŷ이 y를 지나치면 w와 b의 방향도 바꿔준다.

1. 오차와 변화율을 곱하여 가중치 업데이트하기<br>
x[0]일 때 w의 변화율과 b의 변화율에 오차를 곱한다.

```python
err = y[0] - y_hat
w_new = w + w_rate * err
b_new = b + 1 * err
print(w_new, b_new)
```

2. 두 번째 샘플을 이용하여 오차를 구하고 새로운 가중치와 절편 구하기<br>
w_rate = 샘플값 이므로 w_rate = x[1]

```python
y_hat = x[1] * w_new + b_new
err = y[1] - y_hat
w_rate = x[1]
w_new = w_new + w_rate * err
b_new = b_new + 1 * err
print(w_new, b_new)
```

3. 전체 샘플을 반복하기

```python
for x_i, y_i in zip(x, y):
  y_hat = x_i * w + b
  err = y_i - y_hat
  w_rate = x_i
  w = w + w_rate * err
  b = b + 1 * err

print(w, b)
```

4. 얻어낸 모델이 전체 데이터 세트를 잘 표현하는지 그래프를 그려 확인

```python
plt.scatter(x, y)
pt1 = (-0.1, -0.1 * w + b)
pt2 = (0.15, 0.15 * w + b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

5. 여러 에포크를 반복하기<br>
에포크(epoch): 전체 훈련 데이터를 모두 이용해 한 단위의 작업을 진행하는 것<br>
100번의 에포크를 반복하면서 직선이 어떻게 이동하는 지 확인

```python
for i in range(1, 100):
  for x_i, y_i in zip(x, y):
    y_hat = x_i * w + b
    err = y_i - y_hat
    w_rate = x_i
    w = w + w_rate * err
    b = b + 1 * err

print(w, b)
```

```python
plt.scatter(x, y)
pt1 = (-0.1, -0.1 * w + b)
pt2 = (0.15, 0.15 * w + b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

6. 모델로 예측하기<br>
위에서 얻은 모델(ŷ = 913.6x + 123.4)에 새로운 데이터가 발생했을 경우, 모델의 x값에 해당 데이터를 넣고 계산하기만 하면 된다.

```python
x_new = 0.18
y_pred = x_new * w + b
print(y_pred)
```

```python
plt.scatter(x, y)
plt.scatter(x_new, y_pred)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Neuron 클래스 만들기


1. 뉴런 클래스 생성

```python
class Neuron:

	def __init__(self):
		self.w = 1.0
		self.b = 1.0

	def forpass(self, x):
		y_hat = x * self.w + self.b
		return y_hat

	def backprop(self, x, err):
		w_grad = x * err
		b_grad = 1 * err
		return w_grad, b_grad

	def fit(self, x, y, epochs=100):
		for i in range(epochs):
			for x_i, y_i in zip(x, y):
				y_hat = self.forpass(x_i)
				err  = -(y_i - y_hat)
				w_grad, b_grad = self.backprop(x_i, err)
				self.w -= w_grad
				self.b -= b_grad
```

2. `fit()` 메서드에 입력 데이터(x)와 타깃 데이터(y) 전달

```python
neuron = Neuron()
neuron.fit(x, y)
```

3. 학습 완료된 모델의 가중치와 절편 확인하기

```python
plt.scatter(x, y)
pt1 = (-0.1, -0.1 * neuron.w + neuron.b)
pt2 = (0.15, 0.15 * neuron.w + neuron.b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
