# 사이킷런 (Scikit-learn)

- 파이썬의 머신러닝 라이브러리
- 데이터셋과 머신러닝 알고리즘을 포함함
    - It features various classification, resgression and clustering algorithms including support vector machines, random forests, gradient boosting, k-means and DBSCAN
- 파이썬 수치과학 라이브러리와 상호작용함 (NumPy 및 SciPy)

## scikit-learn algorithm cheat-sheet

- 머신러닝의 대표 라이브러리인 만큼 이에 필요한 기술을 포함하는 도표
- **어떤 상황에서 어떤 모델을 만들어야할지** 판단할 수 있는 중요한 지표임
- 머신러닝과 딥러닝에 있어 굉장한 노하우를 포함하니, 자주 찾아올 것

### Problem

- 크게 네 가지 문제를 다룸 (최근 0.18.0 버전부터 인공 신경망(Neural Network)이 본격적으로 추가됨)
- 각각 분류, 회귀, 군집화, 차원축소
    - 분류는 결과를 T/F와 같이 두 가지, 혹은 여러 가지 종류로 구분하는 문제를 의미
    - 회귀는 연속된 수치 값을 예측하는 문제를 의미
    - 군집화는 데이터 간의 유사도를 통해 관계를 파악하는 문제를 의미
        - 즉 group info를 의미있는 sub group, cluster로 조직함
    - 차원축소는 subspace로 데이터를 압축하는 문제를 의미. 이 때 관련있는 정보를 대부분 유지해야
- 군집화와 차원축소를 통한 데이터를 분류와 회귀 문제에 태울 수 있음
    - 군집화는 데이터를 수치화, 벡터화 했을 때 벡터가 유사한 것끼리 묶어줌
    - 차원축소는 계산과 시각화를 원활히 하기 위해 데이터 손실을 감수하고 사용함
    - 이렇듯 군집화를 통한 데이터 묶음과 차원축소를 통한 축소 데이터를 분류와 회귀 문제에 다시 태움

### Work flow

- 도표의 start부터 시작함
- 샘플 데이터의 개수를 질문
    - 50개 미만이면 more data. 충분한 데이터가 필수로 요구됨
- 카테고리를 질문
    - labeled data / quantity
    - labeled data에서 충분한 데이터가 있을 경우 분류 문제, 그렇지 않을 경우 군집화 문제
    - quantity predicting에서 충분한 데이터가 있을 경우 회귀 문제, 그렇지 않을 경우 차원축소 문제
- 이후 다음의 질문들을 만나면서 적당한 모델을 찾아감
    - Number of categories known
    - Few features should be important
    - Just looking - Predict Structure
    - Text data
    - 샘플 수
- Tough luck,,, 에 빠지진 않길
